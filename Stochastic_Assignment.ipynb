```python
import numpy as np
```

### a) Generating the Uniform Random Variables

```python
X = np.random.uniform(0,10,1000)
print(X[0:5])
```

### b) Calculating pdf and cdf

```python
cdf = np.zeros(100)
for i in range(100):
  for j in range(1000):
    if X[j] <= (i-25)/5:    #We see which numbers in the distribution are less, hence the P[X<x] = F_x(x)
      cdf[i] = cdf[i] + 1
cdf = cdf/1000
```

```python
pdf = np.zeros(99)
for i in range(99):
  pdf[i] = (cdf[i+1]-cdf[i])/0.2    #We find the pdf as the derivative of the cdf
```

### c) Plotting the pdf and cdf

```python
from matplotlib import pyplot as plt
x = np.arange(-5,15,0.2)
plt.xlabel('x')
plt.ylabel('F_x(x)')
plt.title('CDF')
plt.plot(x,cdf)
```

```python
plt.xlabel('x')
plt.ylabel('f_x(x)')
plt.title('PDF')
plt.plot(x[0:99], pdf)

#The plot is not exactly like a typical uniform distribution, as this is not the perfect uniform distribution
```
